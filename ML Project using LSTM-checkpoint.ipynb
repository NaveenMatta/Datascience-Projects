# Splitting the dataset

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = scaler.fit_transform(df)
```

# Prepare the data for LSTM model

```python
# Prepare the data for LSTM model
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0]) # if time_step=3 ; i=0,1,2,3 fourth element should be added to the list
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

time_step = 100  # Consider 100 days of historical data for predictions
X, Y = create_dataset(df_scaled, time_step)

# Reshape X to be [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Check shapes of X and Y
print(f"Shape of X: {X.shape}")
print(f"Shape of Y: {Y.shape}")
```

# LSTM Model Creation

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
model.fit(X, Y, batch_size=64, epochs=150, verbose=1)

# Save the model (optional)
model.save('stock_prediction_model.h5')
```

```python
def predict_future_prices(model, data, scaler, days_to_predict):
    predictions = []
    last_data = data[-time_step:]  # Get the last time_step days of data

    for _ in range(days_to_predict):
        input_data = last_data.reshape((1, time_step, 1))
        predicted_price = model.predict(input_data)
        predictions.append(predicted_price[0][0])
        last_data = np.append(last_data[1:], predicted_price)

    return scaler.inverse_transform(np.array(predictions).reshape(-1, 1))
```
